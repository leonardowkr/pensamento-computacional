# **Desafio Cinematográfico**
Seu desafio é desenvolver um programa em Python para registrar os
filmes favoritos e as notas que você dá a cada um deles.

## O Desafio

Seu programa deve:

Solicitar nomes de filmes:  
O usuário informa o nome de um filme e, em seguida, é perguntado
se deseja cadastrar outro.
Armazenar os dados:  
Cada nome de filme deve ser gravado em um array com capacidade
para até 1000 filmes.
O programa deve impedir a inclusão de filmes repetidos.

## Desafio: Cadastro de Filmes e Notas

Registrar avaliações:  
Para cada filme informado, o usuário também deverá inserir uma
nota entre 0.0 e 10.0.
Essas notas serão armazenadas em um array paralelo.

Exibir resultados:  
Ao final, quando o usuário decidir parar de cadastrar filmes, o
programa deverá:
Listar todos os filmes cadastrados.
Calcular e exibir a nota média de todos os filmes.

```python
def media (filmes):
  cnt = len(filmes)
  if cnt == 0:
    return 0
  soma = 0
  for filme in filmes:
    soma +=filme["nota"]
    return soma/cnt

def imprime_filmes(filmes):
  for filme in filmes:
    print(f"O filme {filme['nome']} recebeu a nota {filme['nota']} da crítica!")

def confere_duplicidade (novo, filmes):
  for filme in filmes:
    if novo == filme["nome"]:
      return True
  return False

def mostra_catalogo(filmes):
  print("") #pula linha
  for filme in filmes:
    print(filme['nome']) #mostra opções de filmes
  print("") #pula linha

filmes = []


for i in range (1000):
  operacao ='Selecione uma operação: '
  operacao +='\n ad para adicionar filme'
  operacao +='\n rm para apagar filme'
  operacao +='\n ed para editar flime'
  operacao +='\n sa para sair'
  print(operacao)
  operacao = input()

  match(operacao):
    case 'ad':
      nome = input("\nDigite o nome do filme para adicionar: ")
      if(i > 0):
        if confere_duplicidade(nome, filmes):
          while True:
            print("O filme ", nome, " já foi cadastrado.\n")
            nome = input("Digite outro nome de filme. Para sair, digite 's' ")
            if confere_duplicidade(nome, filmes) == False:
              break
            if(nome == 's'):
              break
      if(nome != 's'): # sair do menu adicionar filme
        nota = float(input("Digite a nota do filme: "))
        print("")
        filmes.append({"nome" : nome, "nota" : nota})

    case 'rm':
      mostra_catalogo(filmes)
      nome = input("Selecione o nome de um filme a cima para remover.")
      token = 0 # # verificador de filme no catalogo
      for filme in filmes:
        if nome == filme["nome"]:
          filmes.remove(filme) #remove filme
          token = 1 # valida que há o filme no catálogo
          print('\nO filme ', nome, ' foi removido com sucesso.\n')
      if token == 0:
            print(f'O filme ', nome, ' não foi encontrado no catálogo.\n')

    case 'ed':
      mostra_catalogo(filmes)
      nome = input("Selecione o nome de um filme a cima para editar: ")
      token = 0 # verificador de filme no catalogo
      for filme in filmes:
        if(nome == filme['nome']): # verifica nome do filme no catalogo
          operacao = input("Para alterar o nome do filme, digite 'filme'. Para alterar a nota, digite 'nota'")
          token = 1 # valida que há o filme no catálogo
      if token == 0:
            print(f'O filme ', nome, ' não foi encontrado no catálogo.\n')

      match(operacao):
        case 'filme':
          for filme in filmes:
            if(filme['nome'] == nome):
              filme["nome"] = input("Insira o nome do filme para edição.")
              print('Filme alterado para: ', filme["nome"])
              print('') #pula linha

        case 'nota':
          for filme in filmes:
            if(filme['nome'] == nome):
              print(f"\nO filme {filme['nome']} recebeu nota {filme['nota']}.")
              nota_editada = input(f"Insira a nova nota do filme {filme['nome']}: ")
              filme['nota'] = nota_editada
              print(f"A nota do filme {filme['nome']} foi alterada para: ", nota_editada)
              print("") # pula linha

    case 'sa':
      print('\nVocê saiu.\n')
      imprime_filmes(filmes)
      break
```
